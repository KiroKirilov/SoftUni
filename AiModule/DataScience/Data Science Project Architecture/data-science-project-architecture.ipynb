```python
times = []
for test in range(51):
  a = np.random.uniform(1, 10000, size = test * 20000)
  b = np.random.uniform(1, 10000, size = test * 20000)
  start = time.time()
  a.dot(b) # Also: np.dot() on numpy arrays
  stop = time.time()
  times.append((stop - start))
plt.scatter(range(len(times)), times)


times = []
for test in range(51):
  a = np.random.uniform(1, 10000, size = test * 20000).tolist()
  b = np.random.uniform(1, 10000, size = test * 20000).tolist()
  start = time.time()
  np.array(a).dot(np.array(b)) # Also: np.dot() on numpy arrays
  stop = time.time()
  times.append((stop - start))
plt.scatter(range(len(times)), times)

```
