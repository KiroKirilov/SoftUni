```python
# dataset has NA for namibia with no quotes so it's interpreted as NaN
tb.loc[tb.iso2.isna(), "iso2"] = "NA"
```

```python
tb
```

```python
tb.describe().T
```

```python
tb.head()
```

```python
tb_tidy = tb.melt(id_vars=["iso2", "year"], var_name="sex_and_age", value_name='cases')
```

```python
tb_tidy
```

```python
tb_tidy['sex'] = tb_tidy.sex_and_age.str.slice(0, 1)
tb_tidy['age_group'] = tb_tidy.sex_and_age.str.slice(1)
```

```python
tb_tidy = tb_tidy.drop(columns=['sex_and_age'])
tb_tidy
```

```python
def process_age_group(age_group):
  ages = {"04": "0-4", "65": "65+", "u": "unknown"}
  if age_group in ages:
    return ages[age_group]
  else:
  # Put a dash before the last two digits
    return f"{age_group[:-2]}-{age_group[-2:]}"
```

```python
tb_tidy.age_group = tb_tidy.age_group.apply(process_age_group)
```

```python
tb_tidy = tb_tidy[["iso2", "year", "sex", "age_group", "cases"]]
tb_tidy = tb_tidy.sort_values(['iso2', 'year'])
```

```python
tb_tidy
```

```python
weather_data = pd.read_csv("./data/weather.csv")
weather_data
```

```python
tidy_weather_data = weather_data.melt(id_vars=["id", "year", "month", 'element'], var_name='day')
tidy_weather_data.day = tidy_weather_data.day.str.slice(1).astype(int)
tidy_weather_data
```

```python
tidy_weather_data = tidy_weather_data.dropna()
tidy_weather_data['date'] = pd.to_datetime(tidy_weather_data[['year', 'month', 'day']])
tidy_weather_data = tidy_weather_data.drop(columns=['year', 'month', 'day'])
```

```python
tidy_weather_data
```

```python
tidy_weather_data = tidy_weather_data.pivot_table(index=['id', 'date'], columns='element', values='value')
```

```python
tidy_weather_data = tidy_weather_data.reset_index()
tidy_weather_data
```

```python
tidy_weather_data.to_csv('./data/weather_tidy.csv', index= None)
```
